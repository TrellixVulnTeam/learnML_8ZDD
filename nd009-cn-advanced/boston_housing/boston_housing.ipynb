### 编程练习3：定义衡量标准
如果不能对模型的训练和测试的表现进行量化地评估，我们就很难衡量模型的好坏。通常我们会定义一些衡量标准，这些标准可以通过对某些误差或者拟合程度的计算来得到。在这个项目中，你将通过运算[*决定系数*](http://stattrek.com/statistics/dictionary.aspx?definition=coefficient_of_determination) R<sup>2</sup> 来量化模型的表现。模型的决定系数是回归分析中十分常用的统计信息，经常被当作衡量模型预测能力好坏的标准。

R<sup>2</sup>的数值范围从0至1，表示**目标变量**的预测值和实际值之间的相关程度平方的百分比。一个模型的R<sup>2</sup> 值为0还不如直接用**平均值**来预测效果好；而一个R<sup>2</sup> 值为1的模型则可以对目标变量进行完美的预测。从0至1之间的数值，则表示该模型中目标变量中有百分之多少能够用**特征**来解释。_模型也可能出现负值的R<sup>2</sup>，这种情况下模型所做预测有时会比直接计算目标变量的平均值差很多。_

在下方代码的 `performance_metric` 函数中，你要实现：
- 使用 `sklearn.metrics` 中的 [`r2_score`](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html) 来计算 `y_true` 和 `y_predict`的R<sup>2</sup>值，作为对其表现的评判。
- 将他们的表现评分储存到`score`变量中。

或 

- (可选) 不使用任何外部库，参考[决定系数的定义](https://en.wikipedia.org/wiki/Coefficient_of_determination)进行计算，这也可以帮助你更好的理解决定系数在什么情况下等于0或等于1。

```python
# TODO 3

# 提示： 导入r2_score
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """计算并返回预测值相比于预测值的分数"""
    
    score = r2_score(y_true, y_predict)

    return score
```

```python
# TODO 3 可选

# 不允许导入任何计算决定系数的库

def performance_metric(y_true, y_predict):
    """计算并返回预测值相比于预测值的分数"""
    y_true1 = np.array(y_true)
    y_predict1 = np.array(y_predict)
    ytmp = np.mean(y_true1)
    SStot = np.sum((y_true1 - ytmp)**2)
    SSreg = np.sum((y_predict1 - ytmp)**2)
    SSres = np.sum((y_true1 - y_predict1)**2)
    
    score = 1.0 - SSres/SStot

    return score
```

### 问题 8 - 回答：
+ K-fold CV是指将训练集平均分为K份，其中K-1份用于模型训练，只有1份进行模型验证。选择不同的K-1份训练集和相对应的不同K种验证集，循环K次训练和验证模型。最后取平均值来衡量模型性能。

+ GridSearchCV中的'cv\_results\_' 罗列出不同参数组合条件下，不同分组的模型训练/验证的评估分数，及其均值和标准差等数据
+ 网格搜索时，如果不使用交叉验证，只使用训练集和测试集。当评估不同参数组合的模型时，因为只使用同一个测试集，调整的参数会在测试集上很好的表现。容易在测试集上出现过度拟合问题，评估指标的泛化性能分数可能不会准确。为了解决这个问题，将训练集拆分出验证集，网格搜索只在训练集上进行，测试集只用于最终的评估。

```python
# TODO 4

#提示: 导入 'KFold' 'DecisionTreeRegressor' 'make_scorer' 'GridSearchCV' 
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV

def fit_model(X, y):
    """ 基于输入数据 [X,y]，利于网格搜索找到最优的决策树模型"""
    
    cross_validator = KFold()
    
    regressor = DecisionTreeRegressor()

    params = {'max_depth':range(1,11)}

    scoring_fnc = make_scorer(performance_metric)

    grid = GridSearchCV(regressor, params,scoring = scoring_fnc,cv = cross_validator)

    # 基于输入数据 [X,y]，进行网格搜索
    grid = grid.fit(X, y)

    #print pd.DataFrame(grid.cv_results_) 
    # 返回网格搜索后的最优模型
    return grid.best_estimator_
```

在下方 `fit_model` 函数中，你需要做的是：

- 遍历参数`‘max_depth’`的可选值 1～10，构造对应模型
- 计算当前模型的交叉验证分数
- 返回最优交叉验证分数对应的模型

```python
# TODO 4 可选

'''
不允许使用 DecisionTreeRegressor 以外的任何 sklearn 库

提示: 你可能需要实现下面的 cross_val_score 函数

def cross_val_score(estimator, X, y, scoring = performance_metric, cv=3):
    """ 返回每组交叉验证的模型分数的数组 """
    scores = [0,0,0]
    return scores
'''

def cross_val_score(estimator, X, y, scoring = performance_metric, cv=3):
    """ 返回每组交叉验证的模型分数的数组 """
    split_tmp = len(X)/cv
    split0_X_train = X[:split_tmp* 2]    
    split0_X_test = X[split_tmp* 2:]
    split0_y_train = y[:split_tmp* 2]
    split0_y_test = y[split_tmp* 2:]
    
    split1_X_train = X[split_tmp:]    
    split1_X_test = X[:split_tmp]
    split1_y_train = y[split_tmp:]
    split1_y_test = y[:split_tmp]
    
    split2_X_train = X[:split_tmp].append(X[split_tmp* 2:])    
    split2_X_test = X[split_tmp:split_tmp* 2]
    split2_y_train = y[:split_tmp].append(y[split_tmp* 2:])
    split2_y_test = y[split_tmp:split_tmp* 2]
    
    estimator0 = estimator.fit(split0_X_train,split0_y_train)
    split0_y_predict = estimator0.predict(split0_X_test)
    split0_score = scoring(split0_y_test,split0_y_predict)
    
    estimator1 = estimator.fit(split1_X_train,split1_y_train)
    split1_y_predict = estimator1.predict(split1_X_test)
    split1_score = scoring(split1_y_test,split1_y_predict)
    
    estimator2 = estimator.fit(split2_X_train,split2_y_train)
    split2_y_predict = estimator2.predict(split2_X_test)
    split2_score = scoring(split2_y_test,split2_y_predict)
    
    
    
    scores = [split0_score,split1_score,split2_score]
    return scores


def fit_model(X, y):
    """ 基于输入数据 [X,y]，利于网格搜索找到最优的决策树模型"""
    if len(X) != len(y):
        return None
    
    tmpBest = 0.0
    tmpBestregressor = None
    for i in range(10):
        regressor = DecisionTreeRegressor(max_depth=i+1)
        tmp_mean_score = np.mean(cross_val_score(regressor, X, y))
        if  tmp_mean_score > tmpBest:
            tmpBest = tmp_mean_score
            tmpBestregressor = regressor
        
    #最优交叉验证分数对应的最优模型
    best_estimator = tmpBestregressor
    
    return best_estimator
```
