```python
import cv2,random
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline

TRAIN_SET_DIR = './train/'
TEST_SET_DIR = './test/'

train_images = [TRAIN_SET_DIR+i for i in os.listdir(TRAIN_SET_DIR)]
test_images = [TEST_SET_DIR+i for i in os.listdir(TEST_SET_DIR)]
train_dogs_images = [TRAIN_SET_DIR+i for i in os.listdir(TRAIN_SET_DIR) if 'dog' in i]
train_cats_images = [TRAIN_SET_DIR+i for i in os.listdir(TRAIN_SET_DIR) if 'cat' in i]
```

## 绘制训练集所有图片尺寸散点图

```python
def draw_train_size_scatter_pic():

    train_images_sizes_x = [ cv2.imread(i).shape[0] for i in train_images]
    train_images_sizes_y = [ cv2.imread(i).shape[1] for i in train_images]

    plt.scatter(train_images_sizes_x, train_images_sizes_y,alpha=0.5,edgecolors= 'white')
    plt.title("the size of pictures in train set")
    plt.xlabel('height')
    plt.ylabel('width')
    plt.grid(True)

draw_train_size_scatter_pic()
```

## 绘制测试集所有图片尺寸散点图

```python
def draw_test_size_scatter_pic():

    test_images_sizes_x = [ cv2.imread(i).shape[0] for i in test_images]
    test_images_sizes_y = [ cv2.imread(i).shape[1] for i in test_images]

    plt.scatter(test_images_sizes_x, test_images_sizes_y,alpha=0.5,edgecolors= 'white')
    plt.title("the size of pictures in test set")
    plt.xlabel('height')
    plt.ylabel('width')
    plt.grid(True)
    
draw_test_size_scatter_pic()
```

## 绘制训练集中部分异常点图片

```python
def draw_outlier_pic():

    outliers =  ['dog.8898.jpg','dog.6725.jpg','dog.11299.jpg','dog.1773.jpg','dog.3889.jpg','dog.6475.jpg',
                      'dog.9517.jpg','cat.7377.jpg','cat.3672.jpg','cat.4338.jpg','cat.10712.jpg',
                      'cat.4308.jpg','cat.7564.jpg','cat.2520.jpg','cat.8456.jpg']

    plt.figure("outliers")

    for i in outliers:
        img = Image.open(TRAIN_SET_DIR+i)
        plt.subplot(3, 5,outliers.index(i)+1)
        plt.title(i)
        plt.imshow(img)

draw_outlier_pic()
```

## 绘制训练集和测试集中随机抽取的图片

```python
def draw_random_pic(images_set):

    tmp = random.sample(images_set, 15)
    plt.figure("cats_dogs")

    for i in tmp:
        img = Image.open(i)
        plt.subplot(3, 5,tmp.index(i)+1)
        plt.title(i)
        plt.imshow(img)


draw_random_pic(test_images)
```
